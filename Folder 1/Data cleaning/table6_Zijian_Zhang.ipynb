```python
firstdf = firstdf.melt(id_vars=["Sort\norder","Major area, region, country or area of destination","Notes","Country code","Type of data (a)"],var_name=["temp"],value_name="Estimated refugee stock at mid-year")
firstdf
```

```python
firstdf = firstdf.assign(Gender = lambda x : x.temp.str[0], Year = lambda x : x.temp.str[1:])
#if rerun, rerun previous block first (because of pop function)
firstdf.pop("temp")
firstdf = firstdf.replace(to_replace=["0","1","2"],value=["both","male","female"])
firstdf
```

```python
firstdf["Sort\norder"] = firstdf["Sort\norder"].astype('int64')
firstdf["Country code"] = firstdf["Country code"].astype('int64')
firstdf
```

```python
firstdf = firstdf.set_index(["Major area, region, country or area of destination","Notes","Type of data (a)","Estimated refugee stock at mid-year","Gender","Year"])
```

```python
firstMain = pd.DataFrame.from_records(
    columns=["id","Major area, region, country or area of destination","Notes","Type of data (a)","Estimated refugee stock at mid-year","Gender","Year"],
    data = [(a+1,b,c,d,e,f,g) for (a,(b,c,d,e,f,g)) in enumerate(firstdf.index.unique())]
)
firstMain
```
