```python
import pandas as pd
thisdf = pd.read_excel("../project data/UN_MigrantStockTotal_2015.xlsx",sheet_name="Table 3",skiprows=14)
thisdf
```

```python
thisdf = thisdf.rename(columns={"International migrant stock as a percentage of the total population (both sexes)" : "01990", "Unnamed: 6" : "01995","Unnamed: 7" : "02000","Unnamed: 8" : "02005","Unnamed: 9" : "02010","Unnamed: 10":"02015",
"International migrant stock as a percentage of the total population (male)" : "11990","Unnamed: 12":"11995","Unnamed: 13":"12000","Unnamed: 14" : "12005","Unnamed: 15" : "12010", "Unnamed: 16" : "12015",
"International migrant stock as a percentage of the total population (female)":"21990","Unnamed: 18" : "21995","Unnamed: 19":"22000","Unnamed: 20":"22005","Unnamed: 21":"22010","Unnamed: 22":"22015"})
thisdf1 = thisdf[1:]

thisdf1 = thisdf1.melt(id_vars=["Sort\norder","Major area, region, country or area of destination","Notes","Country code","Type of data (a)"],var_name=["temp"],value_name="International migrant stock as a percentage of the total population")
thisdf1
```

```python
thisdf1 = thisdf1.assign(Gender = lambda x : x.temp.str[0], Year = lambda x : x.temp.str[1:])
#if rerun, rerun previous block first (because of pop function)
thisdf1.pop("temp")
thisdf1 = thisdf1.replace(to_replace=["0","1","2"],value=["both","male","female"])
thisdf1
```

```python
thisdf1["Sort\norder"] = thisdf1["Sort\norder"].astype('int64')
thisdf1["Country code"] = thisdf1["Country code"].astype('int64')
thisdf1
```

```python
thisdf1.info()
```

```python
thisdf1 = thisdf1.set_index(["Major area, region, country or area of destination","Notes","Type of data (a)","International migrant stock as a percentage of the total population","Gender","Year"])
```

```python
Maintable = pd.DataFrame.from_records(
    columns=["id","Major area, region, country or area of destination","Notes","Type of data (a)","International migrant stock as a percentage of the total population","Gender","Year"],
    data = [(a+1,b,c,d,e,f,g) for (a,(b,c,d,e,f,g)) in enumerate(thisdf1.index.unique())]
)
```

```python
#can not run multiple times because already inplace
Maintable = Maintable.replace(to_replace="..",value=pd.NA)
```

```python
Maintable
```

```python
subtable = thisdf1[["Sort\norder","Country code"]].copy()
subtable["id"] = Maintable.set_index(["Major area, region, country or area of destination","Notes","Type of data (a)","International migrant stock as a percentage of the total population","Gender","Year"])

subtable = subtable.reset_index(drop=True).set_index("id")
subtable
```
