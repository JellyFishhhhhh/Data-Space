```python
import pandas as pd 
thisdf = pd.read_excel("../project data/UN_MigrantStockTotal_2015.xlsx",sheet_name="Table 2",skiprows=14)
thisdf
```

```python
thisdf = thisdf.rename(columns={"Total population of both sexes at mid-year (thousands)" : "01990", "Unnamed: 5" : "01995","Unnamed: 6" : "02000","Unnamed: 7" : "02005","Unnamed: 8" : "02010","Unnamed: 9":"02015",
"Total male population at mid-year (thousands)" : "11990","Unnamed: 11":"11995","Unnamed: 12":"12000","Unnamed: 13" : "12005","Unnamed: 14" : "12010", "Unnamed: 15" : "12015",
"Total female population at mid-year (thousands)":"21990","Unnamed: 17" : "21995","Unnamed: 18":"22000","Unnamed: 19":"22005","Unnamed: 20":"22010","Unnamed: 21":"22015"})
thisdf1 = thisdf[1:]

thisdf1 = thisdf1.melt(id_vars=["Sort\norder","Major area, region, country or area of destination","Notes","Country code"],var_name=["temp"],value_name="Total population at mid-year")
thisdf1
```

```python
thisdf1["Sort\norder"] = thisdf1["Sort\norder"].astype('int64')
thisdf1["Country code"] = thisdf1["Country code"].astype('int64')
thisdf1
```

```python
thisdf1 = thisdf1.assign(Gender = lambda x: x.temp.str[0], Year = lambda x: x.temp.str[1:])
#if rerun, rerun previous block first (because of pop function)
thisdf1.pop('temp')
thisdf1 = thisdf1.replace(to_replace=["0","1","2"],value=["both","male","female"])
thisdf1
```

```python
thisdf1.info()
```

```python
thisdf1 = thisdf1.set_index(["Major area, region, country or area of destination","Notes","Total population at mid-year","Gender","Year"])

Maintable = pd.DataFrame.from_records(
    columns=["id","Major area, region, country or area of destination","Notes","Total population at mid-year","Gender","Year"],
    data = [(a+1,b,c,d,e,f) for (a,(b,c,d,e,f)) in enumerate(thisdf1.index.unique())]
)
```
