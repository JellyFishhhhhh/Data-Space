```python
import pandas as pd
import numpy as np
thisdf = pd.read_excel("../project data/UN_MigrantStockTotal_2015.xlsx",sheet_name="Table 4",skiprows=14)
thisdf
```

```python
thisdf = thisdf.rename(columns={"Female migrants as a percentage of the international migrant stock" : "21990", "Unnamed: 6" : "21995","Unnamed: 7" : "22000","Unnamed: 8" : "22005","Unnamed: 9" : "22010","Unnamed: 10":"22015"})
thisdf1 = thisdf[1:]
thisdf1
```

```python
thisdf1 = thisdf1.melt(id_vars=["Sort\norder","Major area, region, country or area of destination","Notes","Country code","Type of data (a)"],var_name=["temp"],value_name="Percentage of the international migrant stock")
thisdf1
```

```python
thisdf1 = thisdf1.assign(Gender = lambda x : x.temp.str[0], Year = lambda x : x.temp.str[1:])
#if rerun, rerun previous block first (because of pop function)
thisdf1.pop("temp")
thisdf1 = thisdf1.replace(to_replace=["0","1","2"],value=["both","male","female"])
thisdf1
```

```python
thisdf1["Sort\norder"] = thisdf1["Sort\norder"].astype('int64')
thisdf1["Country code"] = thisdf1["Country code"].astype('int64')
thisdf4 = thisdf1
thisdf1
```

```python
thisdf1.info()
```

```python
thisdf1 = thisdf1.replace(to_replace="..",value=0)
```

```python
thisdf1["Percentage of the international migrant stock"] = thisdf1["Percentage of the international migrant stock"].astype(float)
```

```python
thisdf1 = thisdf1.set_index(["Major area, region, country or area of destination","Notes","Country code","Type of data (a)","Percentage of the international migrant stock","Gender","Year"])
```

```python
thisdf2 = pd.DataFrame.from_records(
    columns=["id","Major area, region, country or area of destination","Notes","Country code","Type of data (a)","Percentage of the international migrant stock","Gender","Year"],
    data = [(a,b,c,d,e,100-f,"male",h) for (a,(b,c,d,e,f,g,h)) in enumerate(thisdf1.index.unique())]
)
```

```python
thisdf2
```

```python
thisdf2["Percentage of the international migrant stock"] = thisdf2["Percentage of the international migrant stock"].replace(to_replace=100.0,value=np.nan)
thisdf2["Percentage of the international migrant stock"] = thisdf2["Percentage of the international migrant stock"].astype(str)
```

```python
thisdf2.info()
```

```python
thisdf3 = pd.DataFrame.from_records(
    columns=["id","Major area, region, country or area of destination","Notes","Country code","Type of data (a)","Percentage of the international migrant stock","Gender","Year"],
    data = [(a,b,c,d,e,"100","both",h) for (a,(b,c,d,e,f,g,h)) in enumerate(thisdf1.index.unique())]
)
```

```python
thisdf3
```

```python
thisdf5 = thisdf3.merge(thisdf2,how="outer").merge(thisdf4,how="outer")
thisdf5.pop("id")
thisdf5.pop("Sort\norder")
thisdf5.pop("Country code")
```

```python
thisdf5 = thisdf5.replace(to_replace="..",value=np.nan)
#thisdf5[thisdf5["Gender"] == "female"].iloc[1:30]
```
