```python
import pandas as pd
import numpy as np
thisdf = pd.read_excel("../project data/UN_MigrantStockTotal_2015.xlsx",sheet_name="Table 5",skiprows=14)
thisdf
```

```python
thisdf = thisdf.rename(columns={"Annual rate of change of the migrant stock (both sexes)" : "01995", "Unnamed: 6" : "02000","Unnamed: 7" : "02005","Unnamed: 8" : "02010","Unnamed: 9" : "02015",
"Annual rate of change of the migrant stock (male)" : "11995","Unnamed: 11":"12000","Unnamed: 12":"12005","Unnamed: 13" : "12010","Unnamed: 14" : "102015",
"Annual rate of change of the migrant stock (female)":"21995","Unnamed: 16" : "22000","Unnamed: 17":"22005","Unnamed: 18":"22010","Unnamed: 19":"22015"})
thisdf1 = thisdf[1:]
thisdf1.insert(5,"01990",pd.NA)
thisdf1.insert(11,"11990",pd.NA)
thisdf1.insert(17,"21990",pd.NA)
```

```python
thisdf1
```

```python
thisdf1 = thisdf1.melt(id_vars=["Sort\norder","Major area, region, country or area of destination","Notes","Country code","Type of data (a)"],var_name=["temp"],value_name="Annual rate of change of the migrant stock")
thisdf1
```

```python
thisdf1 = thisdf1.assign(Gender = lambda x : x.temp.str[0], Year = lambda x : x.temp.str[1:])
#if rerun, rerun previous block first (because of pop function)
thisdf1.pop("temp")
thisdf1 = thisdf1.replace(to_replace=["0","1","2"],value=["both","male","female"])
thisdf1
```

```python
thisdf1 = thisdf1.set_index(["Major area, region, country or area of destination","Notes","Type of data (a)","Annual rate of change of the migrant stock","Gender","Year"])
```

```python
Maintable = pd.DataFrame.from_records(
    columns=["id","Major area, region, country or area of destination","Notes","Type of data (a)","Annual rate of change of the migrant stock","Gender","Year"],
    data = [(a+1,b,c,d,e,f,g) for (a,(b,c,d,e,f,g)) in enumerate(thisdf1.index.unique())]
)
```

```python
#can not run multiple times because already inplace
Maintable = Maintable.replace(to_replace="..",value=pd.NA)
```

```python
Maintable = Maintable.set_index("id")
Maintable
```
