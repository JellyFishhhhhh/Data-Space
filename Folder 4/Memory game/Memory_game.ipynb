```python
import pygame
import random

# Initialize Pygame
pygame.init()

# Set up the screen
screen_width = 640
screen_height = 480
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("Memory Game")

# Load images
back_image = pygame.image.load("back.png")
card_images = []
for i in range(1, 9):
    card_images.append(pygame.image.load(f"card{i}.png"))

# Define variables
num_cards = 16
card_width = 100
card_height = 100
cards = []
card_values = []
selected = []
matched = []

# Define functions
def new_game():
    global card_values, selected, matched
    card_values = [i//2 + 1 for i in range(num_cards)]
    random.shuffle(card_values)
    selected = []
    matched = []
    for i in range(num_cards):
        x = i % 4
        y = i // 4
        left = x * (card_width + 10) + 50
        top = y * (card_height + 10) + 50
        cards.append(pygame.Rect(left, top, card_width, card_height))

def draw_screen():
    global matched
    screen.fill((255, 255, 255))
    for i, card in enumerate(cards):
        if i in matched:
            continue
        if i in selected:
            pygame.draw.rect(screen, (255, 0, 0), card, 3)
            card_surf = pygame.Surface((card_width, card_height))
            card_surf.blit(card_images[card_values[i]-1], (0, 0))
            screen.blit(card_surf, card)
        else:
            pygame.draw.rect(screen, (0, 0, 0), card, 3)
            screen.blit(back_image, card)
    pygame.display.flip()

def handle_mouse_click(pos):
    global selected, matched
    for i, card in enumerate(cards):
        if card.collidepoint(pos) and i not in matched and i not in selected:
            selected.append(i)
            if len(selected) == 2:
                if card_values[selected[0]] == card_values[selected[1]]:
                    matched.extend(selected)
                selected = []
            break

# Start a new game
new_game()

# Main game loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.MOUSEBUTTONDOWN:
            pos = pygame.mouse.get_pos()
            handle_mouse_click(pos)
            draw_screen()

    # Check if the game has been won
    if len(matched) == num_cards:
        print("You win!")
        running = False

# Quit Pygame
pygame.quit()
```
